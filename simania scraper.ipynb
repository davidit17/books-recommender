```python
import requests
from bs4 import BeautifulSoup
import pandas
import time
import numpy as np
```

```python
base_url = 'https://simania.co.il/bookdetails.php?item_id='

start_time = time.time()

l=[]
#for page in range(0,1966986,1)
for page in range(965986,966987,1):
    url = base_url+str(page)
    r=requests.get(url)
    c=r.content
    soup=BeautifulSoup(c,"html.parser")
    
    b={}
    
    all = soup.find_all("div")
    
    inf = all[0].find_all("div",{"class":"showrating"})
    
    try : 
        rating = float(inf[0].find_all("b")[1].text)
    except : 
        continue
        
    if rating > 0 :
        b["rating"]=rating
        b["boys"]=inf[1].find_all("b")[1].text
        b["girls"]=inf[1].find_all("b")[2].text
        try : 
            b["avg_age"]=inf[1].find_all("b")[3].text
        except :
            b["avg_age"]=inf[1].find_all("b")[2].text
        b["interest"]=inf[2].find_all("b")[0].text
    

        info = all[0].find_all("div",{"class":"when"})
    
        try:
            b["year"]=info[0].find_all("b")[0].text
        except:
            b["year"]=0
        try:
            b["pages"]=info[0].find_all("b")[1].text
        except:
            b["pages"]=0
        try:
            b["mol"]=info[0].find_all("a")[0].text
        except:
            b["mol"]='unknowen'
        
        b["title"]=soup.find_all("span")[0].text
        try :
            b["author"]=soup.find_all("h3")[0].get_text(strip=True)
        except :
            b["author"]='Unknown'
        
        #b["url"] = page
        l.append(b)

elapsed_time = time.time() - start_time
print("runtime " + str(elapsed_time))
```

```python
df=pandas.DataFrame(l)
```
