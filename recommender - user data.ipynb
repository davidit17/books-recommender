```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix
```

```python
books = pd.read_csv('users.csv',index_col=0 , sep='\t', encoding='utf-8')
```

```python
books.head()
```

```python
books.user_id.unique()
```

```python
# check how many diffrent ratings for a book 

a = books[['rating','name','author']].groupby(['name','author']).count().sort_values(by='rating',ascending = False).head(15).reset_index()
a
```

```python
# drop na

print(books.shape)
books = books.dropna()
print(books.shape)
```

```python
#drop duplicates 

books = books.drop_duplicates()


# Convert dataframe to  user-item matrix

matrix = books.pivot(index='name', columns='user_id', values='rating').fillna(0)


# create sparse matrix

matrix_sparse = csr_matrix(matrix.values)


```

```python
matrix
```

```python
matrix_sparse
```

# KNN


```python

model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')

model_knn.fit(matrix_sparse)
```

```python
# select random book as a base for the recommendation

name_index = np.random.choice(matrix.shape[0])

# enter book name 

book_name = input('book name :')

# if book exists in table , it is the recommendation subject , if not stays random book

if book_name in matrix.index[:,] : 
    name_index = matrix.index.get_loc(book_name) 

name_index
```

```python
# users rating for the chosen book

matrix.iloc[name_index, :]
```

```python
# predict 

distances, indices = model_knn.kneighbors(matrix.iloc[name_index, :].values.reshape(1, -1), n_neighbors = 6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print ('Recommendations for {0}:\n'.format(matrix.index[name_index]))
    else:
        print ('{0}: {1}'.format(i, matrix.index[indices.flatten()[i]]))
```
