```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('books.csv',index_col=0 , sep='\t', encoding='utf-8')
```

```python
books.iloc[[4,50,51,52,53]]
```

```python
# new column of total votes : boys + girls

books['votes'] = books['boys']+books['girls']


# create weighted rating column based on num of votes 

votes = books['votes']
R = books['rating'] 
mean_rating = books.rating.mean()  # mean rating of all books 
minimum_votes = 10 

books['Weighted_Rating']= (R*votes/(votes+minimum_votes)+(mean_rating* minimum_votes/(votes+minimum_votes)))

# men women interest 

books['gender_bias'] = books['girls']/books['votes'] # % girls review of all reviews
```

```python
cols = ['title','author','rating','Weighted_Rating','mol','year','pages','interest','avg_age','boys','girls','votes','gender_bias']

books = books[cols]
```

```python
books.sort_values(by='Weighted_Rating', ascending=False)
```

```python


def find_book(number = 1 ,min_rating=None , min_age=None, max_age=None , mol=None, min_year=None, max_pages=None, 
              popularity=None ,min_votes=None ,gender = None,author=None):
    
    df = books
    
    if min_rating !=None : df = df[df['rating'] > min_rating]
    
    if min_age !=None : df = df[df['avg_age'] > min_age]
        
    if max_age !=None : df = df[df['avg_age'] < max_age]
        
    if mol !=None : df = df[df['mol'] == mol]
        
    if author !=None : df = df[df['author'] == author]

    if min_year !=None : df = df[df['year'] > min_year]
     
    if max_pages !=None : df = df[df['pages'] < max_pages]
    
    if popularity !=None : df = df[df['interest'] > popularity]
        
    if min_votes !=None : df = df[df['votes'] > min_votes]
    
    if gender !=None :
        if gender=='M' : 
            df = df[df['gender_bias'] < 0.5] 
        elif gender=='F' :
            df = df[df['gender_bias'] > 0.5] 
        elif gender > 0 :
            df = df[df['gender_bias'] >= gender] 
        else :
            df = df[df['gender_bias'] <= 1 + gender] 
        
        
    
    df = df.sort_values(by='Weighted_Rating', ascending=False).head(number)
    
    return df
    
    
```

```python
find_book(5,gender='M')
```

```python
find_book(5,gender= -0.7)
```

```python
find_book(3,max_pages=300,gender= +0.7)
```

```python
find_book(max_age = 30 )
```

```python
find_book(3 , min_age = 30 )
```

```python
find_book(popularity = 10000 )
```
